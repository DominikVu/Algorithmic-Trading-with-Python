## 1 Getting Started and Importing Necessary Libraries


```python
import numpy as np  # for scientific computing with arrays and matrices
import yfinance as yf
import pandas as pd  # for data manipulation and analysis
import matplotlib.pyplot as plt # data visualization library
import matplotlib.dates as mdates
```

## 2 Importing the data

```python
gld = yf.download('GLD', start='2021-01-01', end='2023-12-31')
day = np.arange(1, len(gld) + 1)
gld["Day"] = day
gld.drop(columns = ["Adj Close", "Volume"], inplace=True)
gld = gld[["Day", "Open", "High", "Low", "Close"]]
gld.head()
```

```python
gld.info()
```

Moving averages

```python
gld_copy = gld.copy()
gld_copy["9-day"] = gld_copy["Close"].rolling(9).mean()
gld_copy["21-day"] = gld_copy["Close"].rolling(21).mean()
gld_copy[9:22]
```

"Signal" column

```python
conditions = [
    gld_copy["9-day"] > gld_copy["21-day"],  # for signal 1
    gld_copy["9-day"] < gld_copy["21-day"]   # for signal -1
]
choices = [1, -1] 

# np.select to apply the conditions and choices
gld_copy["Signal"] = np.select(conditions, choices, default=0)
gld_copy = gld_copy.dropna()
gld_copy.head() 
```

Returns

```python
gld_copy.head()
```

```python
# The `np.log` function computes the natural log of each element in the 'Close' column.
# The `diff` function then calculates the difference between each log value and its predecessor,
# effectively giving us the log return.
gld_copy['Return'] = np.log(gld_copy['Close']).diff()

# This is done by multiplying the 'signal' column with the calculated 'return' column.
# The 'signal' column presumably contains trading signals where the value is 1 for a long position
# and 0 for no position. By multiplying these signals with the 'return', we get the system return,
# which is the return attributed to the trading system's signals.
gld_copy['System_return'] = gld_copy['Signal'] * gld_copy['Return']
gld_copy['Entry'] = gld_copy.Signal.diff()
gld_copy.head()
```

Plot trades on time series

```python
plt.rcParams['figure.figsize'] = (12, 6)
plt.grid(True, alpha=0.3)

# Plot the 'Close', '9-day', and '21-day' moving averages using the last 252 trading days.
plt.plot(gld_copy.iloc[-252:]['Close'], label='GLD', linewidth=2)
plt.plot(gld_copy.iloc[-252:]['9-day'], label='9-day SMA', linestyle='--')
plt.plot(gld_copy.iloc[-252:]['21-day'], label='21-day SMA', linestyle='-.')

# Plot entry signals: green stars for buy signals and red triangles for sell signals.
plt.scatter(gld_copy[-252:].loc[gld_copy['Entry'] == 2]
            .index, gld_copy[-252:].loc[gld_copy['Entry'] == 2]['9-day'], 
            marker='*', color='g', s=100, label='Buy Signal')
plt.scatter(gld_copy[-252:].loc[gld_copy['Entry'] == -2]
            .index, gld_copy[-252:].loc[gld_copy['Entry'] == -2]['21-day'], 
            marker='v', color='r', s=100, label='Sell Signal')

plt.title('GLD Price and Moving Averages with Trading Signals')
plt.xlabel('Date')
plt.ylabel('Price')

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())

plt.xticks(rotation=45)

plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```
